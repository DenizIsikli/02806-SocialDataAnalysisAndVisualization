##### What is the point of Figure 7?
Figure 7 provides a structured overview of the design space of narrative visualization. It helps designers systematically understand which combinations of **visual narrative techniques** (e.g., highlighting, transitions) and **narrative structure elements** (e.g., ordering, interactivity, messaging) are commonly used together. This allows a designer to:
- Identify standard patterns and genres
- Compare different storytelling approaches
- Make informed design choices when constructing a narrative visualization :contentReference[oaicite:0]{index=0}

##### Most common design choices in each category
From Figure 7 and the analysis:

**Visual Narrative**
- *Visual Structuring*: Consistent visual platform, progress bars, and establishing shots are widely used
- *Highlighting*: Color, annotations, and size differences are the most common
- *Transition Guidance*: Animated transitions and object continuity are frequently used

**Narrative Structure**
- *Ordering*: Linear and user-directed paths dominate
- *Interactivity*: Hover interactions, navigation buttons, and filtering are common
- *Messaging*: Captions, annotations, and headlines are the most frequently used

**Rarely used design choices**
- Tacit tutorials
- Stimulating default views
- Comment repetition and strong narrative summaries

These are underused likely because they require more deliberate narrative design and effort, while many visualizations prioritize exploration over storytelling :contentReference[oaicite:1]{index=1}

##### Favorite genre of narrative visualization
My favorite genre is the **interactive slideshow**. It strikes a strong balance between **author-driven storytelling** and **reader-driven exploration**. The author can guide the user step-by-step through key insights, while still allowing interaction within each step. This makes it especially effective for complex datasets where users need guidance but also benefit from exploration.

##### Least favorite genre of narrative visualization
My least favorite genre is the **partitioned poster**. While it presents multiple views simultaneously, it often lacks a clear narrative flow. Without strong guidance, the viewer may struggle to understand the intended story or the relationships between different components.

##### Mapping Week 7 examples to genres
From the visualizations explored in Week 7:

- Interactive maps (Plotly/Folium) → **Drill-down / reader-driven structures**
- Static summaries or simple plots → **Annotated charts**
- Any step-by-step interactive charts → **Interactive slideshow**

Some examples mix genres. For instance:
- A Plotly visualization with annotations combines **annotated chart + interactive exploration**
- A website page combining text and visuals resembles **magazine style + embedded visualization**

This shows that real-world visualizations often hybridize genres rather than fitting into a single category.

### Life after Segel & Heer

To me the Segel & Heer paper is still foundational, that's why we read it. But I want to tell you a bit more about what's happened after it came out. Especially, a useful question for you guys is: *are all genres created equally?* And it turns out not quite.

***For high-profile public-facing stories***, especially in data journalism, the closest thing to a winner is the author-led, scroll-driven longform piece: annotated charts/maps embedded in a narrative, often with light interaction. That is basically Segel and Heer’s "annotated chart" and "slide show" logic reborn for the web as scrollytelling. Multiple later papers and industry accounts describe [scrollytelling](https://ieeexplore.ieee.org/document/8564193) as a prominent online journalism form, especially for complex explanatory stories. You might [find this piece useful for your final project](https://pudding.cool/process/how-to-make-dope-shit-part-3/).

But ***across dataviz more broadly***, the bigger winner is actually simpler, annotated, mostly static or lightly interactive graphics. Segel and Heer already observed that heavy messaging was under-used in interactive graphs; later work suggests the field moved toward clearer author guidance rather than pure free exploration. In practice, that means headlines, captions, annotations, small multiples, and a bit of hover/filtering—enough interactivity to inspect, not enough to abandon the narrative.

What did not dominate is the fully open-ended, reader-driven interactive as the main storytelling vehicle. Those systems are important for exploration and analysis, but for communication to broad audiences the center of gravity shifted toward hybrid forms: the author sets the path, the reader gets limited agency. (That is almost exactly the middle ground Segel and Heer predicted.)

All of the above is good to keep in mind when you choose a story for your final project.

```python
# =========================================================
# Week 8 — Data Story Plots
# =========================================================

import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import folium
from folium.plugins import HeatMap

# =========================================================
# Load + preprocess (standard block)
# =========================================================

df = pd.read_csv("../../dataset/Week2/merged_police_incidents.csv")

df["Incident Datetime"] = pd.to_datetime(df["Incident Datetime"], errors="coerce")
df["Hour"] = df["Incident Datetime"].dt.hour
df["Year"] = df["Incident Datetime"].dt.year

# Focus crimes
focus_crimes = [
    "Larceny Theft",
    "Assault",
    "Vandalism",
    "Vehicle Theft",
    "Burglary",
    "Drug Offense",
    "Robbery",
    "Weapons Offense",
    "Arson"
]

df = df[df["Unified Category"].isin(focus_crimes)]

# =========================================================
# Create figures folder (safe)
# =========================================================

import os
os.makedirs("../../figures", exist_ok=True)

# =========================================================
# 1. STATIC PLOT (PNG)
# =========================================================

hour_counts = (
    df.groupby(["Year", "Hour", "Unified Category"])
    .size()
    .reset_index(name="Count")
)

hour_counts["NormalizedCount"] = (
    hour_counts.groupby(["Year", "Unified Category"])["Count"]
    .transform(lambda x: x / x.sum())
)

hour_counts = hour_counts.rename(columns={"Unified Category": "CrimeType"})

# aggregate across years
static_data = (
    hour_counts.groupby(["Hour", "CrimeType"])["NormalizedCount"]
    .mean()
    .reset_index()
)

pivot = static_data.pivot(index="Hour", columns="CrimeType", values="NormalizedCount")

pivot.plot(figsize=(10,6))

plt.xlabel("Hour of day")
plt.ylabel("Probability")
plt.title("Average hourly distribution of crime types")

plt.tight_layout()
plt.savefig("../../figures/hourly_distribution.png")
plt.close()

# =========================================================
# 2. MAP (Folium heatmap)
# =========================================================

# sample for performance
df_map = df.sample(10000, random_state=42)

m = folium.Map(location=[37.77, -122.42], zoom_start=12.5)

heat_data = df_map[["Latitude", "Longitude"]].dropna().values.tolist()
HeatMap(
    heat_data,
    radius=6,
    blur=8,
    max_zoom=13,
    min_opacity=0.75
).add_to(m)

m.save("../../figures/crime_map.html")

# =========================================================
# 3. INTERACTIVE PLOT (Plotly)
# =========================================================

interactive_data = (
    hour_counts.groupby(["Hour", "CrimeType"])["NormalizedCount"]
    .mean()
    .reset_index()
)

fig = px.line(
    interactive_data,
    x="Hour",
    y="NormalizedCount",
    color="CrimeType",
    title="Interactive hourly crime patterns by crime type"
)

# optional: hide clutter initially (leave 1 visible if you want)
for trace in fig.data:
    trace.visible = "legendonly"

fig.write_html("../../figures/interactive_hourly_trends.html", include_plotlyjs='cdn')

# =========================================================
# DONE
# =========================================================

print("All figures generated in /figures/")
```
