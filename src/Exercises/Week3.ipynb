>  * Next, we want to generate a slightly more complicated graphic. I'm interested to know if there are certain crimes that happen much more in certain neighborhoods than what's typical. Below I describe how to get that plot going:
>    - First, we need to calculate the relative probabilities of seeing each type of crime in the dataset as a whole. Let's call it `P(crime)`.
>    - Next, we calculate that same probability distribution _but for each PD district_, let's call that `P(crime|district)`.
>    - Now we look at the ratio `P(crime|district)/P(crime)`. That ratio is equal to 1 if the crime occurs at the same level within a district as in the city as a whole. If it's greater than one, it means that the crime occurs _more frequently_ within that district. If it's smaller than one, it means that the crime is _rarer within the district in question_ than in the city as a whole.

````python
# --- Step 2.1: The types of crime and how they take place across San Francisco's police districts ---

import pandas as pd

df_merged = pd.read_csv("../../dataset/Week2/merged_police_incidents.csv")
df_merged["Police District"] = df_merged["Police District"].str.upper()
df_merged = df_merged[df_merged["Police District"] != "OUT OF SF"]
df_merged["Incident Datetime"] = pd.to_datetime(df_merged["Incident Datetime"], errors="coerce")

# List police districts
districts = sorted(df_merged["Police District"].dropna().unique())
print("Districts:", districts)

# Total crimes per district
total_by_district = df_merged["Police District"].value_counts()
print("\nTotal crimes per district:\n", total_by_district)

# Personal focus crimes
focus_crimes = [
    "Assault", "Burglary", "Robbery", "Arson",
    "Vandalism", "Vehicle Theft", "Larceny Theft",
    "Drug Offense", "Weapons Offense"
]

df_focus = df_merged[df_merged["Unified Category"].isin(focus_crimes)]

# Focus crimes per district
focus_by_district = df_focus["Police District"].value_counts()
print("\nFocus crimes per district:\n", focus_by_district)

# P(crime)
p_crime = df_focus["Unified Category"].value_counts(normalize=True)

# P(crime | district)
p_crime_given_district = (
    df_focus
    .groupby("Police District")["Unified Category"]
    .value_counts(normalize=True)
    .unstack(fill_value=0)
)

# Ratio = P(crime | district) / P(crime)
ratio = p_crime_given_district.divide(p_crime, axis=1)

# Plot ratios for each district
import matplotlib.pyplot as plt

fig, axes = plt.subplots(5, 2, figsize=(14, 18), sharex=True)
axes = axes.flatten()

for ax, district in zip(axes, ratio.index):
    ratio.loc[district].sort_values().plot(kind="barh", ax=ax)
    ax.set_title(district)
    ax.axvline(1, linestyle="--")

# Remove empty subplot if any
for ax in axes[len(ratio.index):]:
    ax.axis("off")

plt.tight_layout()
plt.show()
````

````python
# --- Step 2.2: How do crime patterns across districts change over time? ---

# Ensure year column exists
df_merged["year"] = df_merged["Incident Datetime"].dt.year

# Filter out invalid district
df = df_merged[df_merged["Police District"] != "OUT OF SF"].copy()

# Personal focus crimes
focus_crimes = [
    "Assault", "Burglary", "Robbery", "Arson",
    "Vandalism", "Vehicle Theft", "Larceny Theft",
    "Drug Offense", "Weapons Offense"
]

df = df[df["Unified Category"].isin(focus_crimes)]

# Split periods
df_early = df[(df["year"] >= 2003) & (df["year"] <= 2008)]
df_late  = df[(df["year"] >= 2020) & (df["year"] <= 2025)]

def compute_ratio(df_period):
    p_crime = df_period["Unified Category"].value_counts(normalize=True)

    p_crime_given_district = (
        df_period
        .groupby("Police District")["Unified Category"]
        .value_counts(normalize=True)
        .unstack(fill_value=0)
    )

    ratio = p_crime_given_district.divide(p_crime, axis=1)
    return ratio

ratio_early = compute_ratio(df_early)
ratio_late = compute_ratio(df_late)

# Align columns (in case some crimes missing in a period)
ratio_early, ratio_late = ratio_early.align(ratio_late, fill_value=0)

# Side-by-side plots
import matplotlib.pyplot as plt

fig, axes = plt.subplots(len(ratio_early), 1, figsize=(10, 30))

for i, district in enumerate(ratio_early.index):
    ax = axes[i]
    
    df_plot = pd.DataFrame({
        "Early": ratio_early.loc[district],
        "Late": ratio_late.loc[district]
    }).sort_index()
    
    df_plot.plot(kind="bar", ax=ax)
    ax.set_title(district)
    ax.axhline(1, linestyle="--")
    ax.set_ylabel("Ratio")

plt.tight_layout()
plt.show()

# Compute "change" as Euclidean distance between ratio vectors
import numpy as np

changes = {}

for district in ratio_early.index:
    v1 = ratio_early.loc[district].values
    v2 = ratio_late.loc[district].values
    changes[district] = np.linalg.norm(v2 - v1)

changes = pd.Series(changes).sort_values(ascending=False)

print("Districts by change:\n", changes)
````

````python
# --- Step 3.1: Anscombe-like datasets analysis ---

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import urllib.request

# URLs
urls = [
    "https://raw.githubusercontent.com/suneman/socialdata2026/main/files/data1.tsv",
    "https://raw.githubusercontent.com/suneman/socialdata2026/main/files/data2.tsv",
    "https://raw.githubusercontent.com/suneman/socialdata2026/main/files/data3.tsv",
    "https://raw.githubusercontent.com/suneman/socialdata2026/main/files/data4.tsv",
]

datasets = []

# Load datasets
for i, url in enumerate(urls):
    df = pd.read_csv(url, sep="\t", header=None, names=["x", "y"])
    datasets.append(df)

# Stats
for i, df in enumerate(datasets):
    x = df["x"].values
    y = df["y"].values

    print(f"\nDataset {i+1}")

    print(f"Mean x: {np.mean(x):.2f}, Mean y: {np.mean(y):.2f}")
    print(f"Var x: {np.var(x):.3f}, Var y: {np.var(y):.3f}")
    print(f"Correlation: {np.corrcoef(x, y)[0,1]:.3f}")

    a, b, _, _, _ = stats.linregress(x, y)
    print(f"Regression: y = {a:.2f}x + {b:.2f}")

# Plot
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = axes.flatten()

for i, df in enumerate(datasets):
    x = df["x"]
    y = df["y"]

    a, b, _, _, _ = stats.linregress(x, y)

    ax = axes[i]
    ax.scatter(x, y)
    ax.plot(x, a*x + b)
    ax.set_title(f"Dataset {i+1}")
    ax.set_xlim(0, 20)
    ax.set_ylim(0, 15)

plt.tight_layout()
plt.show()
````

> * You already computed Pearson correlations for four very different datasets in Exercise 3.1. Based on that experience, give a concrete example of when Pearson correlation is misleading. Write down the mathematical formulation of the Pearson correlation and explain in your own words how it works.
>   - In Exercise 3.1, all four datasets had almost identical Pearson correlations even though their plots looked very different, which shows it can be misleading. For example, a dataset with a strong nonlinear pattern or an outlier can still have the same correlation as a linear dataset. The formula is:  
>     r = cov(x, y) / (σ_x * σ_y).  
>     It measures how much x and y vary together relative to their individual variation. In simple terms, it checks how well the data fits a straight line, but ignores non-linear patterns and can be heavily influenced by outliers.
>
> * What is the difference between a bar-chart and a histogram? Can you think of a case where using one when you should use the other would lead to a wrong conclusion?
>   - A bar chart is used for comparing discrete categories, while a histogram shows the distribution of continuous data by grouping values into bins. If you used a bar chart instead of a histogram for something like crime times during the day, you might treat each time as a separate category and miss the overall distribution pattern, leading to misleading conclusions about when crimes typically happen.
>

````python
# --- Step 5.2: Recreating plots from DAOST using SF crime data — Part 1 ---

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import gaussian_kde

# ----------------------
# JITTER PLOT (time within hour)
# ----------------------

df = df_merged.copy()
df = df.dropna(subset=["Incident Datetime"])

crime = "Assault"
df_c = df[df["Unified Category"] == crime]

df_c = df_c[(df_c["Incident Datetime"] >= "2021-01-01") &
            (df_c["Incident Datetime"] <= "2021-06-30")]

df_hour = df_c[df_c["Incident Datetime"].dt.hour.isin([12,13,14])]

seconds = (
    df_hour["Incident Datetime"].dt.minute * 60 +
    df_hour["Incident Datetime"].dt.second
)

jitter = np.random.uniform(-0.5, 0.5, size=len(seconds))

plt.figure(figsize=(10, 4))
plt.scatter(seconds, jitter, alpha=0.5)
plt.xlabel("Seconds within hour")
plt.yticks([])
plt.title(f"Jitter Plot – {crime} (13:00–14:00)")
plt.show()


# ----------------------
# HISTOGRAMS (latitude + longitude)
# ----------------------

crime1 = "Assault"
crime2 = "Vehicle Theft"

df1 = df[df["Unified Category"] == crime1].dropna(subset=["Latitude"])
df2 = df[df["Unified Category"] == crime2].dropna(subset=["Latitude"])

df1 = df1.sample(5000, random_state=1)
df2 = df2.sample(5000, random_state=1)

counts1, bins1 = np.histogram(df1["Latitude"], bins=50)
counts2, bins2 = np.histogram(df2["Latitude"], bins=50)

plt.figure(figsize=(10, 4))
plt.plot(bins1[:-1], counts1, label=crime1)
plt.plot(bins2[:-1], counts2, label=crime2)
plt.title("Latitude Distribution")
plt.legend()
plt.show()

counts1, bins1 = np.histogram(df1["Longitude"], bins=50)
counts2, bins2 = np.histogram(df2["Longitude"], bins=50)

plt.figure(figsize=(10, 4))
plt.plot(bins1[:-1], counts1, label=crime1)
plt.plot(bins2[:-1], counts2, label=crime2)
plt.title("Longitude Distribution")
plt.legend()
plt.show()


# ----------------------
# KDE + bandwidth variation
# ----------------------

seconds = seconds.dropna()
x_vals = np.linspace(0, 3600, 200)

if len(seconds) > 1:
    kde_default = gaussian_kde(seconds)
    kde_small_bw = gaussian_kde(seconds, bw_method=kde_default.factor / 3)
    kde_large_bw = gaussian_kde(seconds, bw_method=kde_default.factor * 3)

    plt.figure(figsize=(10, 4))
    plt.plot(x_vals, kde_default(x_vals), label="Default")
    plt.plot(x_vals, kde_small_bw(x_vals), label="Small BW")
    plt.plot(x_vals, kde_large_bw(x_vals), label="Large BW")
    plt.legend()
    plt.title("Bandwidth Effect on KDE")
    plt.show()
else:
    print("Not enough data points for KDE")
````
